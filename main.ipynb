```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA
from langchain.vectorstores import FAISS
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.prompts import PromptTemplate
from langchain.text_splitter import CharacterTextSplitter
from langchain.document_loaders import PyPDFLoader
import os

# Step 1: Set up OpenAI API Key
os.environ["OPENAI_API_KEY"] = key # Replace with your OpenAI API key

# Step 2: Load Documents
# Load the PDF file (Bukhari Sharif)
pdf_loader = PyPDFLoader("bukhari.pdf")  # Replace "bukhari.pdf" with your file path
documents = pdf_loader.load()

# Step 3: Split Documents
# Split the loaded documents into manageable chunks
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = text_splitter.split_documents(documents)

# Step 4: Create Embeddings and Vector Store
# Use OpenAI embeddings to create vector representations of the document chunks
embeddings = OpenAIEmbeddings()
vector_store = FAISS.from_documents(docs, embeddings)

# Step 5: Define Retrieval-Augmented QA Chain
# Set up the retriever to fetch relevant document chunks
retriever = vector_store.as_retriever()
llm = ChatOpenAI(temperature=0.0)  # Use low temperature for factual responses

# Custom prompt to ensure answers are based on Hadiths and include necessary details
custom_prompt = PromptTemplate(
    input_variables=["context", "query"],
    template="""
You are a knowledgeable assistant that only provides answers based on authentic Hadiths from Sahih Bukhari.

When answering, you must:
1. Provide the relevant Hadith text.
2. Mention the Hadith number (if available).
3. Include a brief explanation if mentioned in the context.
4. Avoid adding your own opinion or unrelated content.

Context: {context}
Query: {query}

Answer (with Hadith number and explanation):
"""
)

# Define the QA chain using the retriever and the custom prompt
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    return_source_documents=True,
    # chain_type_kwargs={"prompt": custom_prompt},
    input_key="query"  # Specify that "query" is the input key for the chain
)

# Step 6: Ask a Query
query = "Is it permissible to work long hours that affect my prayers?"
response = qa_chain({"query": query})  # Use "query" as the key in the input dictionary

# Step 7: Display Response and Sources
print("Answer:")
print(response["result"])

# Display relevant Hadith excerpts
print("\nRelevant Hadith and Explanation:")
if "source_documents" in response:
    for doc in response["source_documents"]:
        print("Excerpt:", doc.page_content[:500])  # Display the first 500 characters
```
